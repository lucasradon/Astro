```python
#Erdumfang
#E=Abstand der 2. Orte (Alexandria und Syene/Assuan) 790km
#U=Winkel des Schattens (entspricht beim Wechselwinkel zwischen Alexandria und Assuan)
#E/U=alpha/360°
#U=790km*360°/7°
print(790*360/7)
Exakt: 40077km
```

```python
#Erdradius mit Bogenlänge
#(2*pi*r)/360°=790/7° | * 360° /2 /pi
r = (790*360)/(7*2*math.pi)
print(r)
#Exakt: 6370 km
```
